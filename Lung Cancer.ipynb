### Step 3 : Converting Str variable into Binary to read the model

```python

 GENDER =pd.get_dummies(data['GENDER'],drop_first=True)
```

```python
 LUNG_CANCER =pd.get_dummies(data['LUNG_CANCER'],drop_first=True)
```

```python

data = pd.concat([data,GENDER,LUNG_CANCER], axis=1)
```

```python
data.drop(['GENDER','LUNG_CANCER'],axis=1,inplace=True)
```

```python
data.head()
```

```python

cancer = data.rename(columns = {'M':'GENDER', 'YES':'LUNG_CANCER'}, inplace=False)
```

```python
cancer.head()
```

### Step 4: Logistic Regression Model

```python
from sklearn.model_selection import train_test_split
```

```python
X = cancer.drop('LUNG_CANCER' , axis=1)
y = cancer['LUNG_CANCER']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X , y, test_size=0.25,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression(solver='lbfgs', max_iter=3000)
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

### Model Evaluation

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix,accuracy_score
 
x = data.iloc[:, :-1]
y = data.iloc[:,-1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=1)
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
 
predictions = logmodel.predict(x_test)
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
print(accuracy_score(y_test, predictions))
```
